## Exercices for Chapter 2.2 - MovieLens 1M Data Set

```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('../pydata-book-src/ch02/movielens/users.dat', sep='::', header=None, names=unames)
users[:5]
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('../pydata-book-src/ch02/movielens/ratings.dat', sep='::', header=None, names=rnames)
ratings[:5]
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('../pydata-book-src/ch02/movielens/movies.dat', sep='::', header=None, names=mnames)
movies[:5]
```

### Merging all three tables together

```python
data = pd.merge(pd.merge(ratings, users), movies)
data[:5]
```

```python
data.ix[0]
```

#### Mean movie ratings for each film grouped by gender

```python
mean_ratings = data.pivot_table('rating', rows='title', cols='gender', aggfunc='mean')
mean_ratings[:5]
```

Filter down to movies that received at least 250 ratings:

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles[:5]
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings[:5]
```

Let see the top films among females viewers:

```python
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
top_female_ratings[:10]
```

#### Measure rating disagreement

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
sorted_by_diff[:15]
```

```python
# reverse order of rows, take first 15 rows
sorted_by_diff[::-1][:15]
```

```python
# standard deviation of rating grouped by title
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
# filter down to active_titles
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

```python
# order Series by value in descending order
rating_std_by_title.order(ascending=False)[:10]
```
