# Chapter 1. Introduction

### Hello Tornado

In a console run the following line:
> python src/chapter_1/hello.py --port=8001

In a second console run the following *curl* command:
> curl http://localhost:8001/

output:
```
Hello, friendly user!
```

Run the second *curl* command:
> curl http://localhost:8001/?greeting=Salutations

output:
```
Salutations, friendly user!
```

### String Service

In a console run the following line:
> python src/chapter_1/string_service.py --port=8001

In a second console run the following *curl* command:
> curl http://localhost:8001/reverse/stressed

output:
```
desserts
```

Run the second *curl* command:
> curl http://localhost:8001/wrap -d text=Lorem+ipsum+dolor+sit+amet+consectetuer

output:
```
Lorem ipsum dolor sit amet consectetuer
```

Run the third *curl* command:
> curl http://localhost:8001/wrap?width=20 -d text=Lorem+ipsum+dolor+sit+amet+consectetuer

output:
```
Lorem ipsum dolor
sit amet
consectetuer
```
