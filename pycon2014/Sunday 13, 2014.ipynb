### Understanding basic query
- rough guidelines
   - page response time < 100 ms
   - common queries < 10 ms
   - rare queries < 100 ms
- *pg_stat_statements* -> helps finding slow query in the system
- indexes
   - which do I use?
      - B Tree -> usual
      - GIN -> Generelized Inverted Index -> use with multiple values
      - Generalise Search Trees (GIST)
      - K Nearest Neighbors (KNN)
      - Space Partitioned GIST (SP-GIST)
      - VODKA (coming soon) -> space partition index
   - conditional
   - functional
   - conditional and functional
   - create index concurrently
      - roughly 2-3x slower
      - doesn't lock table

### hstore / JSON
- hstore is a key value store inside the database
   - indexes work with GIN and GIST
- JSON is only validation on text today
   - functional index work -> have fun
- jsonb -> binary representation of JSON

### Pooling

   - application/framework layer

### Backups

   - pg_dump

   - the bytes on disk
   - base backup

# PostgreSQL is Web Scale. Really :)
(presented by Hannu Krosing -> first DBA and archtitect at Skype)

### Why people are using NoSQL
- simplicity of getting started

### What NoSQL usually sacrificies
- the stadard ACID stuff
   - Atomicity
   - Consistency -> usually none by definition
   - Isolation
   - Durability -> usually missing or slow

### What is PostgreSQL
- mature RDBMS relational DB management system
- highly concurrent design
- full ACID quarantees

### What you can sacrifice in PostgreSQL
- You can switch OFF some features
   - Atomicity (always ON)
   - Consistency
      - use no FOREIGN KEYS, etc
      - use JSON, XML, hstore data types
   - Isolation (always ON)
   - Durability

### Using it as a key/JSON-value store
- not just a storage layer
- has very performant storage layer
- you can use it as a simple storage

### How PostgresSQL scaling was created at Skype
- first -> horizontal scaling
- after doing it a few times -> invented pl/proxy (procedural language)

### What is pl/proxy?
- embedded language for PostgreSQL with sole purpose of querying other PostgresSQL DB
- it can just proxy calls
- needs to define 3 functions
   - partitioning
   - cluster version
   - cluster config
   
### Second approach: when your clients need tables
- use PostgreSQL foreign data wrappers and Multicorn
- Multicorn is used for sharding
