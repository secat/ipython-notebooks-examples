```python
import gc
data = {'abc': 123}
gc.get_referents(data)
```

### RSS memory
- representative for the system
- coarse measurement
- heap fragmentation
   - example
      - used 2MB / RSS 2MB
      - allocate 8 MB -> used 10 MB / RSS 10 MB
      - release 8.5 MB -> use 1.5 MB / RSS 10 MB
- difficult to exploit
- *memory_profiler* -> go grab it on pypi

```python
import sys
data = {None: b'x' * 10000}
sys.getsizeof(data)
```

# Garbage Collection in Python
(presented by Benjamin Peterson)

### Outline
- how GC works in various Python implementations
- GC semantics subtleties

### Part 1 - Implementation basics
- what is GC in Python?
   - unused objects are finalized and deallocated
   - When?
      - eventually or never
      - not going out of memory is good
- CPython: reference counting
   - every objects has a count
      - *Py_INCREF()*
      - *Py_DECREF()*
   - major flaw of reference counting -> circular reference
   - CPython's cyclic GC
      - detect reference cycle
      - automatically run when issuing the command *gc.collect()*
- PyPy
   - written in RPython
   - has pluggable GC
   - current default GC: *minmark*
   - no need to worry about reference cycles -> do not use reference counting
   - uses a marking scheme
   - the GC pauses
   - has an incremental GC (in PyPy 2.2)
   - summary:
      - pluggable
      - generational
      - incrmental
      - integrated with the JIT
      
### Part 2 - GC semantics
- *\_\_dell\_\_* kills kittens
   - Cycle finalizers:
      - CPython < 3.4 -> give up
      - CPython >= 3.4: PEP 442
- PEP 442 "Safe object finalization"

# Designing Poetic APIs

- programmers write code that humans can understand

### Advice (design smell checklist)
- don't be an architect astronaut
- the best library are extracted, not invented
   - bad code example -> nose plugin progress bar

   - example *get(key, default)* better than *fetch(default, key)*

   - watch for long arg lists
- composability
   - watch for classes with lots of state
   - watch for deep inheritance hierarchies

- safety

# Writing RESTful web services with Flask
(presented by Miguel Grinberg [github](github.com/miguelgrinberg))

### Basic APIs with Flask
- Read resource collections
   - GET requests
   - The response includes the URLs for all the resources in the collection
   - the *get_url()* helper methods in the model returns the URLs for each resource
   - the *to_json()* helper methods in the model class returns the JSON representation as a Python dict
   - succesful response has status code 200 (OK)
- Read individual resources
- per-resource collections
- Create a resource
   - POST requests
   - import JSON from the *from_json()* helper functions
   - a succesful response has a status code 201
- Update a resource
   - PUT requests
   - a succesful has a status code 200 (OK)
- Delete a resource
   - DELETE requests
- Exporting JSON representations -> *to_json()*
   - Flask's *url_for()*
- Importing JSON representations -> *from_json()*
- Error handling
   - exception are caught with Flask's **errorhandler**
- Unit Testing
   - built on top of Flask's own test client
   
### Advanced APIs with Flask
- Authentication
      - very important
   - Flask-HTTPAuth
      - the *auth.verify_pasword* decorators registers a password verification
   - Protecting API routes
      - the *auth.login_required* decorators protects a route
   - Global API protection
      - Flask's **before_request** decorators registers a function that runs before requests
   - Using authentication tokens
      - the **verify_password** decorated function validates token given in the username field
- Automatic JSON responses
   - JSON decorator implementation -> *jsonify*
- Rate limiting
   - *@rate_limit(limit, per) \# maximum of requests per seconds*
   - a request above the limit is returned 429
- Pagination
   - *@paginate()* decorator
   - used for large response
- HTTP caching
   - adding cache directives
      - e.g. *@no_cache*
   - cache control decorator -> *cache_control*
- Conditional requests
   - ETag headers -> *@etag* decorator

# Performance Testing and Profiling: A Virtuous Cycle
(presented by Dan Crosta)

### Performance testing basics
- for web app -> generate requests against your app
- measur time and response rates

### Types of Performance testing
- stress test
   - generate excessive load
   - how much?
   - identify breaking point
- load test -> more recommended
   - generate specific, constant load
   - what if?
   - capacity planning
   
### Best practices
- isolate testing from external influences
   - use dedicated LD environment
   - *scaled down* copies of all components
- generate load consistently
   - random considered harmful
   - automate, automate, ...

### Profiling
- Batteries included
   - *cProfile*, *pstats*
- *pstats* interesting functions
   - *print_stats*
   - *print_callees* <- use it when you find a slow function
   - *print_callers*

### Profiling in practice
- tell us why is it slow
- good to identify un-optimized code
- good for identifying bottlenecks
   - distinguish between slow external resource and slow application code
   
### Other 3rd party profiler
- *line_profiler* (http://pythonhosted.org/line_profiler/)
- *yappi* (https://code.google.com/p/yappi/)

### Instrumentation
- use *statsd* to collect time-series metrics
- two key instruments
   - counters let you know how many thing happend
   - timers let you know how long they each took
- learn what's normal for your app
   - bonus -> alert when things are not normal
- let you answer -> does the real world match expectations

### The virtuous cycle
- instrumentation & alerting
- performance testing
- profiling
- performance optimization
