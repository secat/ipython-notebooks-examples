# Introduction to regular expression

### The re module
- use re.search()
- commonly used methods: group, groups, groupdictt
- if regex does not match -> re.search() returns None

### Python "raw" strings
- prefix a string using r to get a "raw" string

### The Basics


```python
import re
re.search('abc', 'abcdef')
```

```python
re.search('xyz', 'abcdef')
```

### Repetition
- "+" one or more times
- "*" zero or more times
- "?" zero or more one times
- "{x}"

### Character Sets
- enclosed in "[]" characters are called character set

```python
re.search('gr[ae]y', 'gray')
```

```python
re.search('gr[ae]y', 'grey')
```

```python
re.search('[0-9]{3}-[0-9]{4}', '886-7654')
```

- [a-z0-9-]+ will match one or more characters between a and z

### Character Clases
- \w word character
- \d numbers (digit)
- \s whitespace character
- \b word boundary characters

### Negation
- match character that are not suppose to be there

```python
re.search(r'abcd[^d]', 'abc')
```

### Special characters
- '^' matching beginning of string/line
- '$'

### Full match introspection
- use the group -> m = re.search(); m.group()

### Backreferences
- using ()
- when using groups() on the search it will return a tuple

### Named backreferences
- add ?P<foo>
- ex: ^(?P<city>[\w ]+)
- you can use the groupdict() method on this

### Conclusion
- topic not covered: alternation ("or"), amnd others

# Cache me if you can: memcached, caching patterns and best practices
(presented by Guillaume Ardaud)

### Memcached
- key/object store
- O(1) everything
- where does it fit?
  - relational db
  - persistent store
  - RAM store

### Typical API

   cache.set()
   cache.set_many({'lunch': diner})
   var = cache.get_many(['lunch', 'diner'])
   cache.incr()
   cache.decr()
   cache.add()
   cache.clear()
   
### Key names
- ASCII based
- Not crazy long
- but explicit enough
- do not use user input

### How do I list all keys in memcached?
- you don't -> its a cache not a DB
- don't search on StackOverflow

### Some subtleties
- expiration
- if the memory is full -> some data get evicted, if so you get None when accesing it
- it pre-allocate memory in 1 MB pages with chuncks (slab chuncks) -> chunck are seperated in class
- if all pages of the needed class are full -> get a free page and give it the needed class

### Useful command lines
- man memcached is your friend
- memcached -v -> verbose output
- memcached -M: do not evict when no more memory -> send an error instead

### Examples
- to_json -> use a cache to get a pre-generated json representation
- thundering herd
- caching large values:
   - use a 2-phase fetch
   - use paginated cache (create a PaginatedCache class with a set(), get() and delete() methods)

# By your bootstraps: Porting your application to Python3

### Porting strategies
- using 2to3 at a starting point is painfully slow on large code base
- "Straddling" in a single codebase
   - conditional imports
   - use "compatible subset" of Python syntax
   - "six" module can help (but you might not need it)
   
### Targeting Python Versions
- syntax changes make Python2 < 2.6 hard -> so we are ok, since we use 2.7.2
- incompatibilities make Python3 < 3.2 hard
   - PEP 3333 fixes WSGI in Py3k
   - callable() restored in 3.2
- 3.3 restires u'' literals
- 3.2 is "system Python3" on some LTS systems

### Managing Porting Risks
- Ports are great opportunities for bug injection
- Fear of breaking working software IS the major problem, not the effort

### Bottom-up Porting
- port packages with no dependancies first
- then port packages with already-ported dependancies

### "Common subset" idioms
- [Lennart Regebro's book - conversion](http://python3porting.com/noconv.html)
- *python2.7 -3* can point out problem areas
- Distinguish bytes vs text
   - use b'' / u'' for all literals
- adopt the new syntax
   - except ... as ...
   - print() as a function
- use new stdlib facilities
   - e.g. io.Bytes

### Testing avoid hair-loss
- untested code is where the bugs go to hide
- 100% coverage is ideal before porting
   - unit tests -> libraries
   - functional tests -> applications
- Measure coverage -> see PyPi for the coverage module
- work to improve assertions as well as coverage
   - assert contracts, not implementation details
- do not use doctests
- use *tox* -> ensure that tests pass under all supported versions

### Considerations for C Extensions
- [Lennart Regebro's book - C extension](http://python3porting.com/cextensions.html)
- use a python reference implementation
- design the same API as C
- 100% coverage
- run all the same tests on the C version

### Hygeine
- tell/signal in *setup.py* explicitely which versions of Python you support
- consider bumping major version
- apply continuous integration
   - Travis-CI
   - Jenkins
   - Shining Panda for windows

### Resources
- python3porting
- testrun.org/tox/latest
- *six* on PyPi

# What is Async, How dows it work & When should I use it?
(presented by A. Jesse J. Davis)

### Examples
- cpu bound web app (subs example) -> do not use async
   - throughput bound by computation
- normal web app (pizza cook example) -> use async
   - throughput bound by memory
   - clients -> server -> backend
- websocket application
   - number of clients bound by memory
   - clients -> server <- events
   
### C10K 
[paper](kegel.com/c10k.html)

### Why is async hard to code
- need to store state

### ways to store state 
- threads
- callbacks: async application uses callbacks
- and coroutines, greenlets, and so on
   
### So, what is async?
- single-threaded
- i/o concurrency
- non-blocking sockets
- epoll / kqueue
- event loop
   
### asyncio
- AKA *Tulip*
- Python 3.4 standard library
- implements PEP 3156
- Standard event loop
- Coroutines

### chat example layers
- application -> *example.py*
- protocol -> *autobahn websockets*
- transport -> *asyncio*
- event loop -> *asyncio* 
- selectors

### Review
- asyncio uses non-blocking sockets
- event loop tracks sockets, and the callback waiting for them

### should I use it?
- Yes:
   - slow backend
   - websockets
   - many connections
- No:
   - CPU-bound
   - no async driver (e.g. MySQL is not async)
   - no async expertise
