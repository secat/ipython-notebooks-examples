# Athentication

## Twisted Cred: An example

In a first terminal run the **echo credential server**:

> python src/chapter_9/echo_cred.py

In a second terminal run the following command:

> telnet localhost 8000

Here is the output (*with right credentials*):

```
Trying ::1...
telnet: connect to address ::1: Connection refused
Trying 127.0.0.1...
Connected to localhost.
Escape character is '^]'.
user pass
Login successful, please proceed.
Hi
Hi
Quit
Quit
^]
telnet> quit
Connection closed.
```

Run again the same command (*with wrong credentials*):

> telnet localhost 8000

Here is the output:

```
Trying ::1...
telnet: connect to address ::1: Connection refused
Trying 127.0.0.1...
Connected to localhost.
Escape character is '^]'.
foo bar
Login denied, goodbye.
Connection closed by foreign host.

```

## Athentication in Twisted Applications

The plugin application example uses the *twistd* `AuthOptionMixin`. Type the following command in a terminal:

> twistd echo --help-auth

Here is the output:

```
Usage: --auth AuthType[:ArgString]
For detailed help: --help-auth-type AuthType

  AuthType	ArgString format
  ========	================
  memory	A colon-separated list (name:password:...)
  file  	Location of a FilePasswordDB-formatted file.
  unix  	No argstring required.
```

Lets run the **echo credential plugin server** with the following command line:

> twistd -n echo --auth memory:user:pass:foo:bar

Here is the output:

```
2015-01-03 13:56:27-0500 [-] Log opened.
2015-01-03 13:56:28-0500 [-] twistd 14.0.2 (/Users/sophie_serge/work/python/vtwisted/bin/python 2.7.6) starting up.
2015-01-03 13:56:28-0500 [-] reactor class: twisted.internet.selectreactor.SelectReactor.
2015-01-03 13:56:28-0500 [-] EchoFactory starting on 8000
2015-01-03 13:56:28-0500 [-] Starting factory <echo_cred_app.EchoFactory instance at 0x10a52c638>
```

We can execute the same `telnet` experience from last example.
