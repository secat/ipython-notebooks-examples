# Web Servers

## The Structure of an HTTP Request

In a first terminal run the **web echo server**:

> python src/chapter_4/webecho.py

In a second terminal run the following command:

> telnet localhost 8000

Here is the output:

```
Trying 127.0.0.1...
Connected to localhost.
Escape character is '^]'.
```

Write the following lines:

> GET / HTTP/1.1

> Host: localhost:8000

> X-header: "My test header"

> `enter-key`

Here is the output:

```
HTTP/1.1 200 OK

You said:

GET / HTTP/1.1
Host: localhost:8000
X-header: "My test header"
Connection closed by foreign host.
```

In a browser enter the following url:

> localhost:8000

Here is the output from the browser:

