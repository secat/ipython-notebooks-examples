# Deploying Twisted Applications

## The Twisted Application Infrastructure

### TAC File

In a terminal run the following command:

> twistd -y /src.chapter_6/echo_server.tac

The **echo server** from chapter 2 is run as an application daemon.

In the terminal, check the content of the `twistd.pid` file and run the following command:

> kill -15 pid
