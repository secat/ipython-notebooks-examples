# Testing

## Writing and Running Twisted Unit Tests with Trial

Here is an example of how to run unit tests with `trial`:

> trial src/chapter_11/test_foo.py

Here is the output:

```python
from IPython.display import Image
Image(filename='images/trial_test_foo_screenshot.png', width=50, retina=True)
```

It is also possible to run the test by specifying the a test class:

> trial src/chapter_11/test_foo.MyFirstTestCase

And it is possible to run a single class test case:

> trial src/chapter_11/test_foo.MyFirstTestCase.test_something
